This project aims to analyze the relationship between GDP and life expectancy.  It also aims to create visuals that explain this relationship.

```python
#Importing the necessary libraires
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('all_data.csv')
print(data.head())
```

```python
#print(data.dtypes)

print(data.info())  #This is saying that there are no null values so I don't think  need to data clean
print(data.describe(include='all'))   #There is no strings so no data cleaning is needed.
```

```python
#Generating graphs to better explain the data
# plt.figure(1)
# sns.boxplot(data)
# plt.show()
# plt.clf()

fig2 = plt.figure(2)
plt.semilogx(data.GDP, data['Life expectancy at birth (years)'], marker = 'o', linestyle = 'none')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy at Birth (years)')
plt.title('Life Expectancy vs GDP 2000-2015')
plt.show()
#plt.clf()

fig3 = plt.figure(3)
fig, axs = plt.subplots(1,2, num=3)
sns.boxplot(x='Country', y='GDP', data=data, ax=axs[0])
axs[0].set_yscale('log')
axs[0].tick_params(axis='x', rotation=45)
#Changing the x tick labels
current_labels = axs[0].get_xticklabels()
new_labels = [label.get_text() for label in current_labels]
new_labels[new_labels.index('United States of America')] = 'USA'
axs[0].set_xticklabels(new_labels, rotation=45, ha='right')
#Samething but for life_expectancy
sns.boxplot(x='Country', y='Life expectancy at birth (years)', data=data, ax=axs[1])
axs[1].tick_params(axis='x', rotation=45)
#Changing the x tick labels
current_labels = axs[1].get_xticklabels()
new_labels = [label.get_text() for label in current_labels]
new_labels[new_labels.index('United States of America')] = 'USA'
axs[1].set_xticklabels(new_labels, rotation=45, ha='right')
plt.subplots_adjust(wspace=0.4)
plt.show()
#plt.clf()

fig4 = plt.figure(4)
sns.scatterplot(x='Year', y='Life expectancy at birth (years)', data = data, hue = 'Country')
handles, labels = plt.gca().get_legend_handles_labels()

# Modify the label for a specific country (e.g., 'United States of America' to 'USA')
labels = [label if label != 'United States of America' else 'USA' for label in labels]

# Update the legend with the new labels
plt.legend(handles=handles, labels=labels, title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')

#plt.clf()
plt.show()

#Saving figures
fig3.savefig('GDP and life expectancy.png', dpi = 300, bbox_inches='tight')
fig4.savefig('life expectancy vs time.png', dpi = 300, bbox_inches='tight')
```

These graphs show that a very low GDP is asscoiated with a lower life expectancy.  But other than that there is not a huge impact.
